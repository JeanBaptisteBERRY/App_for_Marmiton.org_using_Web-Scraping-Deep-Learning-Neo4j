### Packages

```python
# import des packages de base
import numpy as np

# import du package de traduction
from googletrans import Translator

# import des packages de deeplearning
from keras.models import load_model
from keras.applications import imagenet_utils
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
from keras.applications import VGG16

# import de matplot
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

### Définition du modèle de Deep Learning

```python
def image_ingredient(image_path):
    
    '''Initialisation du traducteur'''
    translator = Translator()
    
    '''Tranformation de l'image en entrée de modèle pour qu'elle puisse être traitée'''
    inputShape = (224, 224)
    image = load_img(image_path, target_size=inputShape)
    image = img_to_array(image)
    image = np.expand_dims(image, axis=0)
    preprocess = imagenet_utils.preprocess_input
    image = preprocess(image)    
    
    '''Chargement des poids pré-entrainés pour le modele VGG16'''
    model = VGG16(weights="imagenet")
    
    '''Récupération de la prédiction pour l'image traitée'''
    preds = model.predict(image)
    
    '''Récupération du label de la prédiction pour l'image traitée'''
    preds_en = imagenet_utils.decode_predictions(preds)[0][0][1]
    
    '''Traduction du label anglais en français pour être utilisé par la base de données Ne04j écrite en français'''
    preds_fr = translator.translate(preds_en, dest='fr')
    preds_fr_word = (str(preds_fr).split('text'))[1].split(',')[0][1:]
    
    '''La fonction renvoie le label en français de l'image'''
    return preds_fr_word
```
